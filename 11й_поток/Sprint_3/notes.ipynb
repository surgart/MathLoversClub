```python
import sympy as sp

def build_matrix(n):
    matrix = []
    for i in range(n):
        row = []
        for j in range(n):
            d = 1 if i == j else 2
            row.append(d)
        matrix.append(row)
    return matrix

A = [
    [-16, 16, 1],
    [-4, 16, 4],
    [1, 16, 16],
]

A = sp.Matrix(A)
A.inv().det(), A.det()
```

```python
from sympy.abc import n

def gen_matrix(n):
    m = []
    for i in range(1, n + 1):
        row = []
        for j in range(1, n + 1):
            if i == j:
                row.append(3)
            elif i == j - 1:
                row.append(2)
            elif i == j + 1:
                row.append(1)
            else:
                row.append(0)
        m.append(row)
    return m

A = sp.Matrix(gen_matrix(3))
A.det()
```
